```python
import numpy as np
import pandas as pd
import math
d=pd.read_excel("CustomerDemographic.xlsx")
```

```python
d.head()
```

```python
d.drop(columns='default',inplace=True)
```

```python
d.head()
```

```python
d['customer_id'].is_unique
```

```python
d.set_index('customer_id',inplace=True)
```

```python
d.head()
```

```python
s=d['gender']
M=s.str.contains('M')
F=s.str.contains('F')
U=s.str.contains('U')
M[:5]
```

```python
d['gender']=np.where(M,'Male',np.where(F,'Female',np.where(U,'Male','NaN')))
```

```python
d.head()
```

```python
d['job_title'].isnull().sum()
```

```python
d['job_industry_category'].isnull().sum()
```

```python
d.drop(columns='job_industry_category',inplace=True)
```

```python
d.head()
```

```python
Nan=d['job_title'].str.contains("NaN")
d['job_title']=d['job_title'].fillna('Unknown')
d['last_name']=d['last_name'].fillna(' ')
```
