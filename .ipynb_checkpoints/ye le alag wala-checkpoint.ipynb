Lets start with filling missing value
- DOB
- job_title

```python
df['DOB']=pd.to_datetime(df['DOB'])
```

```python
# now our dates are converted into datetime object lets make a new columns as birth year
df['birth_year']=df['DOB'].dt.year
df.head()
```

```python
# dropping the DOB columns
df.drop('DOB',axis=1,inplace=True)
df.head()
```

```python
print(df['birth_year'].isnull().sum())
df['Age']=2020-df['birth_year']
df.drop('birth_year',axis=1,inplace=True)
# lets find out teh age
```

```python
df.head()
```

```python
# lets drop the name and last name of people to clean data also drop defaults and job_title as it have large null values
df.drop(['first_name','last_name','job_title','default'],axis=1,inplace=True)
df.head()# now our dataset look like a bit clear
```

```python
df['gender'].value_counts()
```

AS we can see that in gender column there are male female and U (others) as well as some typos like F M Femal lets process it

Convert female as 0 
male as 1
U as 2

```python
def process(x):
    if x=='Female' or x =='Femal' or x =='F':
        return 0
    elif x=='Male' or x=='M':
        return 1
    else :
        return 2
    
```

```python
df['gender']=df['gender'].apply(process)
```

```python
df.dropna(subset=['job_industry_category'], how='all', inplace=True)
```

```python
df['job_industry_category'].isnull().sum()
```

```python
df['tenure'].isnull().sum()
```

```python
#lets fill tenure by the mean of that
df['tenure'].fillna(value=df['tenure'].mean(), inplace=True)
```

```python
df['tenure'].isnull().sum()
```

```python
df['Age'].isnull().sum()
#lets fill tenure by the mean of that
df['Age'].fillna(value=df['Age'].mean(), inplace=True)
```
